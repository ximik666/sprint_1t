```python
salary_split = data['Salary Estimate'].str.split('-', expand=True)
data["salary_min"] = salary_split[0].str.replace('\D','')
data["salary_max"] = salary_split[1].str.replace('\D', '')
data["salary_max"] = pd.to_numeric(data["salary_max"])
data["salary_min"] = pd.to_numeric(data["salary_min"])
data["salary_mean"] = (data["salary_min"] + data["salary_max"])/2
data.head()
```

Построить график компаний с самым большим кол-вом вакансий

```python
data['Company Name'] = data['Company Name'].str.replace('\n.*','')
data_company_group = data.groupby("Company Name")["Job Title"].count().reset_index().sort_values(by = "Job Title", ascending=False).head(20)
data_company_group
```

```python
data_company_group.set_index("Company Name",inplace=True)
data_company_group['Job Title'].plot(kind="bar",)
```

```python
from wordcloud import WordCloud
import stopwords
stop_words = stopwords.get_stopwords("en")
stop_words = set(stop_words)
all_words_list = data["Job Description"].to_list()
total_words = []
for lst in all_words_list:
    for prep in [",",":",":","!","?",".", "(",")"]:
        lst = lst.replace(prep," ").lower().replace("   "," ").replace("  "," ")
        for word in lst.split():
            if word not in stop_words:
                total_words.append(word)
string_words = " ".join(total_words)



wordcloud = WordCloud(width = 1920, height = 1200,max_words = 2000,stopwords=stop_words,min_word_length = 3, colormap = "Reds",prefer_horizontal = 0.5,background_color = "white").generate(string_words)
plt.imshow(wordcloud, interpolation='bilinear')
```
